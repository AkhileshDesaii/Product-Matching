```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#loading Flipkart data
df_fk = pd.read_csv("flipkart_com-ecommerce_sample.csv")
```

```python
df_fk.head()
```

```python
df_fk.columns
```

```python
#dropping/removing the unnecessary columns
remove = ['crawl_timestamp', 'product_url', 'product_category_tree', 'pid', 'image', 'is_FK_Advantage_product', 'description', 'product_rating', 'overall_rating', 'brand', 'product_specifications']
df_fk = df_fk[df_fk.columns.difference(remove)]
```

```python
#renaming of Flipkart data columns for further understanding
df_fk = df_fk.rename(columns = {"product_name":"Product_name_in_Flipkart"})
df_fk = df_fk.rename(columns = {"retail_price":"Retail_price_in_Flipkart"})
df_fk = df_fk.rename(columns = {"discounted_price":"Discounted_price_in_Flipkart"})
```

```python
df_fk.head()
```

```python
#loading Amazon data
df_amz = pd.read_csv("amz_com-ecommerce_sample.csv", encoding='latin1')
```

```python
df_amz.head()
```

```python
df_amz.columns
```

```python
#dropping/removing the unnecessary columns
remove = ['crawl_timestamp', 'product_url', 'product_category_tree', 'pid', 'image', 'is_FK_Advantage_product', 'description', 'product_rating', 'overall_rating', 'brand', 'product_specifications']
df_amz = df_amz[df_amz.columns.difference(remove)]
```

```python
#renaming of Amazon data columns for further understanding
df_amz = df_amz.rename(columns = {"product_name":"Product_name_in_Amazon"})
df_amz = df_amz.rename(columns = {"retail_price":"Retail_price_in_Amazon"})
df_amz = df_amz.rename(columns = {"discounted_price":"Discounted_price_in_Amazon"})
```

```python
df_amz.head()
```

```python
#Merging two data tables for analysing
df = pd.merge(df_fk, df_amz, on='uniq_id')
print(df)
```

```python
#final data after merging
df.head()
```

```python
#Checking for Retail price difference between Amazon and Flipkart
df['Retail_price_diff'] = df['Retail_price_in_Amazon'] - df['Retail_price_in_Flipkart']
```

```python
df.Retail_price_diff.head()
```

```python
#Checking for discounted price difference between Amazon and Flipkart
df['discounted_price_diff'] = df['Discounted_price_in_Amazon'] - df['Discounted_price_in_Flipkart']
```

```python
df.discounted_price_diff.head()
```

# As we can observe Retail price is more in Flipkart than Amazon and also Amazon had given more discount than Flipkart

# So overall Amazon Product costs are less compared to Flipkart
